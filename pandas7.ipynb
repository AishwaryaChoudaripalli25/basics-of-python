### aggregate

```python
tips.groupby('day').agg({'tip':np.mean,'day':np.size})
#agg=aggregation,np.mean indicates the avg tip on a particular day
#size indicate the count of people on particular day
```

```python
tips.groupby('day').agg({'tip':np.sum,'day':np.size})
#np.sum-> sum of all avg tips
```

```python
51.96/19
```

### groupby and size same column

```python
tips.groupby(['day','smoker']).agg({'tip':[np.size,np.mean]})
```

```python
tips.groupby(['day','sex']).agg({'tip':[np.size,np.sum,np.mean,np.amin,np.amax]})
#np.min->min tip by female or male
#np.max->max tip given by female or male
```
