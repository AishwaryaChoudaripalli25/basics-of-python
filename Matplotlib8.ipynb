### Pie Chart – autopct with one decimal point 


```python
hours=[8,10,2,4]
activity=['sleep','work','eat','play']
col=['lightgreen','lightpink','cyan','lightyellow']
plt.pie(hours,labels=activity,colors=col,startangle=90,shadow=True,explode=(0,0,0,0.1),autopct='%1.1d%%')
plt.title('pie chart')
plt.show()
```

### Pie Chart – autopct and Lebel examples  

```python
hours=[5,7,3,4]
activity=['sleep','work','eat','play']
col=['tomato','palevioletred','cadetblue','darkseagreen']
plt.subplot(224)#outpct with three decumal point
plt.pie(hours,labels=activity,colors=col,shadow=True,autopct='%1.2d%%')
plt.axis('equal')
plt.legend(loc='upper left')
plt.subplot(222)#outpct with integer
plt.pie(hours,labels=activity,colors=col,shadow=True,autopct='%d%%')
plt.title('pie chart')
plt.axis('equal')#equal aspect ensures that pie drawn as a circle
plt.legend(loc='upper right')
plt.show()
```

### multiple graphs

```python
x1=np.arange(10,30,2)
y1=[1,5,6,7,8,3,4,5,6,8]
x2=np.arange(10,30,2)
y2=[5,4,2,8,3,6,9,5,2,7]

plt.subplot(211)
plt.plot(x1,y1)
plt.title('graph 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(212)
plt.plot(x2,y2)
plt.title('graph 2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

### pading

```python
x1=np.arange(10,30,2)
y1=[1,5,6,7,8,3,4,5,6,8]
x2=np.arange(10,30,2)
y2=[5,4,2,8,3,6,9,5,2,7]

plt.subplot(211)
plt.plot(x1,y1)
plt.title('graph 1')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.subplot(212)
plt.plot(x2,y2)
plt.title('graph 2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.tight_layout(pad=1,h_pad=1,w_pad=1)
plt.show()
```
