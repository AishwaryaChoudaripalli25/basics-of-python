### supervised learning

```python
import seaborn as sns
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

### load dataset and make a copy

```python
tips=sns.load_dataset('tips')
tips1=tips
print(tips1)
```

### Create Box plot to check outliers


```python
sns.boxplot(data=tips1,x='day',y='total_bill')
plt.show()
```

```python
tips[tips1['day']=='Fri']
```

```python
tips[tips1['day']=='Fri'].describe()
```

### Create dis plot 


```python
sns.displot(tips[tips1['day']=='Fri'],kde=True)
plt.show()
```

```python
sns.displot(tips[tips1['day']=='Fri']['total_bill'],kde=True)
plt.show()
```

### Apply log Transformation to address outliers 


```python
tips1['total_bill']=np.log10(tips1['total_bill'])
print(tips1)
```

```python
sns.boxplot(data=tips1,x='day',y='total_bill')
plt.show()
#impact of outlier has reduced
```

```python
sns.displot(tips[tips1['day']=='Fri']['total_bill'],kde=True)
plt.show()
#impact of outlier has addressed
```

```python
sns.displot(tips1['total_bill'],kde=True)
plt.show()
```
